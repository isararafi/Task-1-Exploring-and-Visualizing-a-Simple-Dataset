```python
# Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# For prettier plots
sns.set(style="whitegrid")

# Show plots inside the notebook
%matplotlib inline
```

```python
iris = sns.load_dataset('iris')
```

```python
# Shape of the dataset
print("Shape of the dataset:", iris.shape)

# Column names
print("\nColumn names:", iris.columns.tolist())

# First few rows
iris.head()
```

```python
# Info
print("\nDataset Info:")
print(iris.info())

# Summary Statistics
print("\nSummary Statistics:")
iris.describe()
```

```python
# Scatter plot: Sepal Length vs Sepal Width colored by species
plt.figure(figsize=(8, 6))
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species')
plt.title('Sepal Length vs Sepal Width')
plt.show()
```

```python
# Histograms for each feature
iris.hist(figsize=(10, 8))
plt.suptitle('Histograms for All Features')
plt.show()
```

```python
# Better individual histograms with KDE
for column in iris.columns[:-1]:  # Skip 'species'
    plt.figure(figsize=(6, 4))
    sns.histplot(data=iris, x=column, kde=True)
    plt.title(f'Distribution of {column}')
    plt.show()
```

```python
# Box plot for each feature (all species together)
for column in iris.columns[:-1]:
    plt.figure(figsize=(6, 4))
    sns.boxplot(y=iris[column])
    plt.title(f'Box plot of {column}')
    plt.show()
```

```python
# Box plot by species
for column in iris.columns[:-1]:
    plt.figure(figsize=(6, 4))
    sns.boxplot(data=iris, x='species', y=column)
    plt.title(f'Box plot of {column} by species')
    plt.show()
```
